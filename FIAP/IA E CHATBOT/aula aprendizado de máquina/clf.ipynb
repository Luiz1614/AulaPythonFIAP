```python
df= pd.read_csv("dados_flu.csv")
df.sample(5)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.drop(["seguro_de_saúde", "seguro_de_saúde.1", "indústria_de_emprego", "ocupação"], axis=1)
```

```python
df = df.dropna()
```

```python
plt.figure(figsize=(18,18))
sns.heatmap(df.corr(), cmap="coolwarm", center=0, annot=True)
```

```python
X = df[["opinião_sobre_o_risco_sazonal",
        "recomendação_médica_sazonal",
        "comportamento_lavagem_das_mãos"]]

y = df["vacina_sazonal"]
```

```python
!pip install scikit-learn 
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
vai_vacinar = DecisionTreeClassifier()
vai_vacinar.fit(X, y)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(vai_vacinar, fontsize=12)
```
